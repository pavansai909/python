```python
#Which cities were the most used during business and home hours

merged.info()
 
# we realise that the date colum it an onject and so we have to convert it to a date time datatype
```

```python
# converting the column DATATIME from object datatype to datetime datatype

merged["DATETIME"]= pd.to_datetime(merged["DATETIME"])
merged.info()

#The column DATETIME datatype ha snow been changed
```

```python
#At this point we can now find the usage in the there days between office hours and home hours 

# Lets start by looking at working hours..

working_hrs = merged[(merged['DATETIME'].dt.hour >= 8) & (merged['DATETIME'].dt.hour <= 17)] ["PRODUCT"].value_counts()
working_hrs

#from the output we realise that no product was used used during the buisness hours.
#so this means that its either people dont use their phones during the business hours or,
```

```python
#now we can look at home hours

after_work = merged[(merged['DATETIME'].dt.hour >= 17) & (merged['DATETIME'].dt.hour <= 24)] ["PRODUCT"].value_counts()
before_work= merged[(merged['DATETIME'].dt.hour >= 0) & (merged['DATETIME'].dt.hour < 8)] ["PRODUCT"].value_counts()

home_hours = after_work + before_work
home_hours

#we realise that each product was used during this hours which means
#1. the date collected may have been for the homehours only or in this cities people use there phones during home hours only
#2. the results also show that the most product that is used is sms so the company can thinkmof investing more on sms
```
